```python
import whisper
import speech_recognition as sr
from gtts import gTTS
import os
import requests
import json
import playsound

# Load Whisper model
model = whisper.load_model("large")

# Google Gemini API setup
GEMINI_API_KEY = "AIzaSyBXEe4EabD2igigZX9jFjjHk0hh35I6DB8"
GEMINI_URL = "https://generativelanguage.googleapis.com/v1beta/models/gemini-pro:generateContent"

# Step 1: Record audio and convert to text
def record_and_transcribe():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("ðŸŽ™ï¸ Speak now...")
        audio = recognizer.listen(source)

        with open("temp.wav", "wb") as f:
            f.write(audio.get_wav_data())

        result = model.transcribe("temp.wav")
        return result['text']

# Step 2: Send query to Gemini API
def query_gemini(prompt):
    headers = {
        "Content-Type": "application/json"
    }
    params = {
        "key": GEMINI_API_KEY
    }
    data = {
        "contents": [
            {"parts": [{"text": prompt}]}
        ],
        "generationConfig": {
            "temperature": 0.7,
            "maxOutputTokens": 150,
            "topP": 1,
            "topK": 1
        }
    }

    response = requests.post(GEMINI_URL, params=params, headers=headers, data=json.dumps(data))
    if response.status_code == 200:
        return response.json()['candidates'][0]['content']['parts'][0]['text']
    else:
        return "Sorry, I couldn't fetch the answer."

# Step 3: Convert answer to speech and play it
def speak_text(text):
    tts = gTTS(text)
    tts.save("response.mp3")
    playsound.playsound("response.mp3")

# Run the full pipeline
question = record_and_transcribe()
print("ðŸ§  You asked:", question)

answer = query_gemini(f"Give a compact and clear response to: {question}")
print("ðŸ’¡ Gemini Response:", answer)

speak_text(answer)
```
